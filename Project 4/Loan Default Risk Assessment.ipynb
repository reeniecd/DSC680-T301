The machine learning model I have developed serves the purpose of determining specific criteria. The primary objective of this project is to construct predictive models that assess the likelihood of a potential borrower defaulting on a loan. These models utilize information gathered from borrowers during the application process, such as age, income, and salary, along with financial data, credit history, and debt-to-income attributes obtained from the Credit Bureau.

Additionally, supplementary models will be created to segment customers based on demographic and behavioral factors. This segmentation will provide lenders with valuable insights into the personalities and habits of their customers. By gaining a deeper understanding of their customers, lenders can make more informed decisions and tailor their services to better meet the needs and preferences of different customer segments.

##### Correlation Matrix

```python
import seaborn as sns
# Create correlation matrix
corr_matrix = df.corr().abs()

sns.heatmap(corr_matrix)
```

```python
# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
# Drop features 
df.drop(to_drop, axis=1, inplace=True)
df.head()
```

Removing any variable that correlation value greater than 0.95.The dataset features dropped from 30 to 25.

### Target variable

```python
# Create dictionary that map loan status to loan type
loan_type_dict = {
    'Fully Paid': 'Good',
    'Charged Off': 'Bad',
    'Current': 'Good',
    'Default': 'Bad',
    'Late (31-120 days)': 'Bad',
    'In Grace Period': 'Bad',
    'Late (16-30 days)': 'Bad',
    'Does not meet the credit policy. Status:Fully Paid': 'Good',
    'Does not meet the credit policy. Status:Charged Off': 'Bad'
}

loan_type_target = {
    'Fully Paid': 1,
    'Charged Off': 0,
    'Current': 1,
    'Default': 0,
    'Late (31-120 days)': 0,
    'In Grace Period': 0,
    'Late (16-30 days)': 0,
    'Does not meet the credit policy. Status:Fully Paid': 1,
    'Does not meet the credit policy. Status:Charged Off': 0
}

df['loan_type'] = df['loan_status'].map(loan_type_dict)
df['loan_type_target'] = df['loan_status'].map(loan_type_target)
df.drop(['loan_status'], axis=1)
df.head()
```

The loans with the statuses Current and Fully Paid are definitely considered "Good".  Charged Off , Default, Late, In-Grace-period will be considered "Bad".
